```python
# The PDB processing on pandas dataframe is based on the following documents. 
#http://rasbt.github.io/biopandas/tutorials/Working_with_PDB_Structures_in_DataFrames/

import os
from biopandas.pdb import PandasPdb
```

```python
def process_pdb(path):
    for pdb_file in os.listdir(path):
        i=1
        pdb_code = pdb_file[:30]
        PDB_path = os.path.join(path,pdb_file)
        ppdb = PandasPdb()
        PDB_Read = ppdb.read_pdb(PDB_path)
        PDB_DF = PDB_Read.df['ATOM']
        PDB_drop_col = PDB_DF.drop(['blank_1', 'alt_loc', 'blank_2', 'insertion', 'blank_3', 'blank_4', 'segment_id'],
                       axis = 1)
        PDB_drop_col.to_csv(f"{pdb_code}_generic_renumber.csv")
        i +=1
```

```python
path = '/set/the/path/to/pdb/file'
```

```python
process_pdb(path)
```
